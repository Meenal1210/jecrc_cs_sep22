```python
import matplotlib.pyplot as plt
import numpy as np
```

# Types of Graph
1. plot -          line graph, between two parameters 
2. bar -           bar graph of single parameter (optional - categorical data)
3. hist -          histogram of a bunch of data i.e.frequency of single parameters 
4. scatter -       dot graph between two parameters 
5. boxplot/box -   statistical overview in form of box for a single parameter 
                   mean , outliers, quantities (25% , 75%)
                   (optional - categorical data)
6. stem            more like bar graphs, Lines instead of bars
7. violinplot -    violin shaped statistical overview graph
                   spead of data (more like hist)
                   single parameter (optional cat-data)

```python
plt.plot(x,y)
plt.title('PLOT GRRAPH')
plt.show()

plt.bar(x,y)
plt.title('BAR GRRAPH')
plt.show()

plt.hist(x)
plt.title('HISTOGRAM')
plt.show()

plt.scatter(x,y)
plt.title('Scatter')
plt.grid()
plt.xlabel('x -- >')
plt.ylabel('y -- >')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()


plt.boxplot(y)
plt.title('BOXPLOT')
plt.show()

plt.stem(x,y)
plt.title('STEM')
plt.show()

plt.violinplot(x)
plt.title('VIOLINPLOT')
plt.show()
```

```python
# to print  a comparison plot
y2 = x**4
```

```python
fig = plt.figure(figsize = (10,5))

plt.plot(x, y, lw=3 , marker='o',markersize=15, markerfacecolor='yellow')
plt.plot(x,y2, marker='v', markersize = 10)
plt.legend(['y=x**3', 'y2=x**4'])
plt.grid()
plt.show()
```
