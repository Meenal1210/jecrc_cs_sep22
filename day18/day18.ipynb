#### Graph Functions:
1. plot                          - Simple Line Plot
2. plot.area                     - Filled Version of Line Plot
3. plot.bar / plot.barh          - Bar Graph
4. plot.box                      - Box Plot
5. plot.density                  - Density Line (Smooth Curved)
6. plot.hexbin                   - Hexagon Version of Scatter with Density
7. plot.hist                     - Histogram
8. plot.kpde                     - Kernel Density Estimator (Desity Graph)
9. plot.pie                      - Pie Chart of Categorical Data
10. plot.scatter                 - Scatter Graph

# Seaborn

```python
import seaborn as sns
```

##### Categories
1. Categorical Plots   (hues can be used | comparison with categorical data)
     a. countplot   (a single parameter)
     b. barplot     (1 categorical parameter & 1 numerical parameter | calculates mean by default)
     c. boxplot     ()
     d. violinplot
     e. swarmplot
2. Distribution Plots
     a. distplot
     b. displot
     c. joinplot
     d. pairplot
     e. kdeplot
     f. rugplot 
     g. scatterplot
3. Matrix Plots
     a. heatmap
     b. clustermap
     c. pairplot (same as above one)
4. Regresion Plot 
     a. lmplot
     b. regplot
5. Grid Plot
     a. PairGrid
     b. JoinGrid
     c.FacetGrid


```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe(include='all')
```

```python
tips['day'].value_counts()
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(y='sex',data=tips)
```

```python
sns.countplot(x='day', data=tips, hue='sex')
```

```python
sns.barplot(x='day',y='total_bill', data=tips)
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day', y='total_bill',data=tips, estimator=sum)
```

```python
sns.barplot(x='day', y='total_bill', data = tips, hue = 'time', estimator = np.mean)
```

```python

sns.boxplot(x='day', y='total_bill', data = tips, hue = 'time')
```

```python
sns.boxplot(x='day', y='total_bill', data = tips, hue = 'smoker')
plt.show()
sns.violinplot(x='day', y='total_bill', data = tips, hue = 'smoker')
plt.show()
sns.swarmplot(x='day', y='total_bill', data = tips, hue = 'smoker')
plt.show()
```
