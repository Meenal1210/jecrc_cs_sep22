```python
import warnings
warnings.filterwarnings('ignore')

# Categorical Data
for col in cat_cols:
    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    sns.countplot(data = data, x=col, palette=choice(p_list))
    plt.title(col)
    plt.subplot(1,2,2)
    data[col].value_counts().plot.pie(autopct='%1.1f%%', figsize=(8,4))
    plt.show()

# Numerical data
for col in num_cols:
    color = [random(), random(), random()]
    fig = plt.figure(figsize=(10,3))
    fig.suptitle(col)
    plt.subplot(1,3,1)
    sns.distplot(data[col], color = color)
    plt.subplot(1,3,2)
    sns.boxplot(data = data, x = col)
    plt.subplot(1,3,3)
    sns.violinplot(data = data, x = col)
    plt.show()
```

### MultiVariate Analysis

```python
for col in cat_cols:
    if col == 'Product':
        continue
for ncol in num_cols:
    plt.figure(figsize=(10,3))
    plt.suptitle(f'{col} vs {ncol}')
    plt.subplot(1,3,1)
    sns.boxplot(data = data, x=col, y=ncol, hue='Product', palette=choice(p_list) )
    plt.subplot(1,3,2)
    sns.violinplot(data = data, x=col, y=ncol, hue='Product', palette=choice(p_list))
    plt.subplot(1,3,3)
    sns.barplot(data = data, x=col, y=ncol, hue='Product', palette=choice(p_list))
    plt.show()
```
